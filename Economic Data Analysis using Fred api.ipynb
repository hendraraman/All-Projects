```python
unemp_data = fred.search("unemployment",filter=("frequency","Monthly"))
```

```python
unemp_data.head()
```

```python
unemp_data = unemp_data.query("seasonal_adjustment == 'Seasonally Adjusted' and units == 'Percent'")
```

```python
unemp_data.head()
```

```python
unemp_data = unemp_data.loc[unemp_data.title.str.contains("Unemployment")]
```

```python
unemp_data.index
```

```python
# getting unemployment rates for every state
all_results = []
for i in unemp_data.index:
    results = fred.get_series(i)
    results = results.to_frame(name = i)
    all_results.append(results)
```

```python
unemp_results = pd.concat(all_results,axis = 1).tail(900) # creating a separate column for each state
```

```python
unemp_results
```

```python
unemp_states = unemp_results.drop("UNRATE",axis = 1)
unemp_states
```

```python
countries = unemp_data.title.str.replace("Unemployment Rate in","")
# countries.to_frame()
countries = countries.to_frame()
countries
# countries = countries.to_dict()
countries = countries.title.apply(lambda x: x if len(x) < 20 else 0)
countries = countries.replace("Unemployment Rate","").to_dict()

unemp_states.columns = [countries[c] for c in unemp_states.columns]
unemp_states = unemp_states.drop(0,axis = 1) # only states
```

```python
unemp_states
```

```python
px.line(unemp_states)
```

```python
# unemp rates of all states during april 2020
rates_at_april = unemp_states.loc[unemp_states.index == '2020-04-01'].T.dropna()
```
